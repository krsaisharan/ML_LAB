```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from google.colab import files

uploaded = files.upload()
diabetes = pd.read_csv("Diabetes.csv")

cols = ['Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']
diabetes[cols] = diabetes[cols].replace(0, np.nan)

for c in cols:
    diabetes[c] = diabetes[c].fillna(diabetes[c].median())

num_cols = diabetes.select_dtypes(include=np.number).columns[:-1]

for c in num_cols:
    Q1 = diabetes[c].quantile(0.25)
    Q3 = diabetes[c].quantile(0.75)
    IQR = Q3 - Q1
    diabetes = diabetes[(diabetes[c] >= Q1 - 1.5*IQR) & (diabetes[c] <= Q3 + 1.5*IQR)]

mm = MinMaxScaler()
diabetes[num_cols] = mm.fit_transform(diabetes[num_cols])

ss = StandardScaler()
diabetes[num_cols] = ss.fit_transform(diabetes[num_cols])

print("Preprocessing completed")
diabetes.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from google.colab import files

uploaded = files.upload()
adult = pd.read_csv("adult.csv")

adult.replace('?', np.nan, inplace=True)

cat_cols = adult.select_dtypes(include='object').columns
num_cols = adult.select_dtypes(include=np.number).columns

for c in cat_cols:
    adult[c] = adult[c].fillna(adult[c].mode()[0])

for c in num_cols:
    adult[c] = adult[c].fillna(adult[c].median())

le = LabelEncoder()
for c in cat_cols:
    adult[c] = le.fit_transform(adult[c])

for c in num_cols:
    Q1 = adult[c].quantile(0.25)
    Q3 = adult[c].quantile(0.75)
    IQR = Q3 - Q1
    adult = adult[(adult[c] >= Q1 - 1.5*IQR) & (adult[c] <= Q3 + 1.5*IQR)]

mm = MinMaxScaler()
adult[num_cols] = mm.fit_transform(adult[num_cols])

ss = StandardScaler()
adult[num_cols] = ss.fit_transform(adult[num_cols])

print("Adult income preprocessing completed")
adult.head()
```
